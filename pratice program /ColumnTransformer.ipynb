```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("tested.csv")
```

```python
df
```

```python
df.drop(["PassengerId","Name","Ticket","Cabin","Embarked"],axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(["Survived"],axis=1),df["Survived"],test_size=0.3,random_state=21)
```

```python
c=ColumnTransformer(transformers=[("Simple",SimpleImputer(),[2]),
        ("Ohe",OneHotEncoder(drop="first",sparse_output=False),[1]),
        ("Stand",StandardScaler(),slice(2,6))],remainder="passthrough")
```

```python
c.fit_transform(x_train)
```
