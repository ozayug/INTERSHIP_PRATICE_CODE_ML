```python
df1['Order Date'] = pd.to_datetime(df1['Order Date'])
```

```python
df1["Order Date"]
```

```python
df1.sort_values('Order Date', inplace=True)
```

```python
revenue_by_date = df1.groupby('Order Date')['Total Revenue'].sum()
```

```python
revenue_by_date
```

```python
plt.figure(figsize=(10, 6))
plt.plot(revenue_by_date.index, revenue_by_date.values, marker='o', linestyle='-')
plt.title('Total Revenue Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Revenue')


```
