```python
df.drop(["Gender","Ethnicity","ParentalEducation","Tutoring","ParentalSupport","Extracurricular","Sports","Music","Volunteering","GradeClass"],axis=1,inplace=True)
```

```python
x=df.drop("Absences",axis=1)
```

```python
y=df["Absences"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=21)
```

```python
s=StandardScaler()
```

```python
x_train_s=s.fit_transform(x_train)
x_test_s=s.fit_transform(x_test)
```

```python
x_train_s
```

```python
c=s.get_feature_names_out()
```

```python
x_train_s=pd.DataFrame(x_train_s,columns=c)
```

```python
np.round(x_train_s.describe(),2)
```

```python
fig,(a1,a2)=plt.subplots(ncols=2,figsize=(12,5))
a1.scatter(x_train["StudyTimeWeekly"],x_train["GPA"])
a1.set_title("Before scalling")
a2.scatter(x_train_s["StudyTimeWeekly"],x_train_s["GPA"],color="black")
a2.set_title("After scalling")
plt.show()
```

```python
m=MaxAbsScaler()
```

```python
x_train_m=m.fit_transform(x_train)
x_test_m=m.fit_transform(x_test)
```

```python
x_train_m
```

```python
h=m.get_feature_names_out()
```

```python
x_test_m=pd.DataFrame(x_train_m,columns=h)
```

```python
np.round(x_test_m.describe(),2)
```

```python
fig,(a1,a2)=plt.subplots(ncols=2,figsize=(12,5))
a1.scatter(x_test["StudyTimeWeekly"],x_test["GPA"])
a1.set_title("Before scalling"),
a2.scatter(x_test_m["StudyTimeWeekly"],x_test_m["GPA"],color="red")
a2.set_title("After scalling")
plt.show()
```
