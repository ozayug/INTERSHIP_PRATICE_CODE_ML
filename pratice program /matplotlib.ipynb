```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.linspace(1,100)
```

```python
x
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y,color="red")
plt.title("wave graph")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.show()
```

```python
x=[10,20,30,40,50]
x1=[60,70,80,90,100]
y=["A","B","C","D","E"]
```

```python
plt.bar(y,x,color="red")
plt.bar(y,x1,color="blue",alpha=0.2)
plt.title("bar_graph")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.show()
```

```python
plt.pie(x,labels=y,autopct='%0.0f%%',shadow=True,explode=(0.3,0,0.1,0,0))
plt.show()
```

```python
plt.scatter(x,y)
```
